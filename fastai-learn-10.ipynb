```python
from fastai.text.all import *
path = untar_data(URLs.IMDB)
```

```python
path
```

```python
files = get_text_files(path, folders = ['train', 'test', 'unsup'])
```

```python
len(files)
```

```python
txt= files[0].open().read()
```

```python
txt[:75]
```

```python
spacy = WordTokenizer()
```

```python
toks = first(spacy([txt]))
```

```python
print(coll_repr(toks,30))
# coll_repr (c, max_n=10):  String repr of up to max_n items of (possibly lazy) collection c
```

```python
tkn = Tokenizer(spacy)
```

```python
print(coll_repr(tkn(txt),31))
```
