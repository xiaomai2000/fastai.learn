```python
from keras.layers import Dense, Dropout, Conv2D, BatchNormalization
from keras.layers import MaxPooling2D, ZeroPadding2D, Flatten
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
model.add(Conv2D(96, (11, 11), strides=(4,4), input_shape=(227,227,3),activation='relu'))
```

```python
model.add(BatchNormalization())
```

```python
model.add(MaxPooling2D((3,3), strides=(2,2)))
```

```python
#
```

```python
model.add(ZeroPadding2D((2,2)))
model.add(Conv2D(256,(5,5), strides=(1,1), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((3,3), strides=(2,2)))
```

```python
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(384, (3,3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(384,(3,3), activation='relu'))
```

```python
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, (3,3), activation='relu'))
model.add(MaxPooling2D((3,3), strides=(2,2)))
```

```python
model.add(Flatten())
model.add(Dense(4096,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4096,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1000,activation='softmax'))
```

```python
model.summary()
```
