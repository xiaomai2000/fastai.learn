```python

from keras.layers import Dense, Dropout, Conv2D, BatchNormalization
from keras.layers import MaxPooling2D, ZeroPadding2D, Flatten, Activation
from keras.models import Sequential

model = Sequential()

# 1
model.add(Conv2D(96, (11, 11), strides=(4,4), input_shape=(227,227,3), padding='valid', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((3,3), strides=(2,2), padding='valid'))


# 2
model.add(Conv2D(256,(5,5), strides=(1,1), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((3,3), strides=(2,2)))


# 3
model.add(Conv2D(384, (3,3), strides=(1,1), padding='same', activation='relu'))
model.add(Conv2D(384,(3,3),strides=(1,1), padding='same', activation='relu'))
model.add(Conv2D(256, (3,3), strides=(1,1), padding='same', activation='relu'))
model.add(MaxPooling2D((3,3), strides=(2,2)))


# 4
model.add(Flatten())
model.add(Dense(4096,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4096,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1000,activation='relu'))
model.add(Dropout(0.5))

#output layer
model.add(Dense(2))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
        optimizer='sgd', metrics=['accuracy'])
model.summary()
```

```python
from PIL import Image
import numpy as np
from keras.utils import to_categorical
```

```python
path = '/root/data/train'
```

```python
train_X = np.empty((2000,227,227,3), dtype="float16")
train_Y = np.empty((2000,),dtype="int")
```

```python
for i in range(1000):
  file_path = path + "/cats/cat."+ str(i) + ".jpg"
  #print(file_path)
  image = Image.open(file_path)
  resized_image = image.resize((227,227),Image.ANTIALIAS)
  img = np.array(resized_image)
  train_X[i,:,:,:] = img
  train_Y[i] = 0
```

```python
for i in range(1000):
  file_path = path + "/dogs/dog."+ str(i) + ".jpg"
  #print(file_path)
  image = Image.open(file_path)
  resized_image = image.resize((227,227),Image.ANTIALIAS)
  img = np.array(resized_image)
  train_X[i+1000,:,:,:] = img
  train_Y[i+1000] = 0
```

```python
train_X /= 255
train_Y = to_categorical(train_Y,2)
```

```python
index = np.arange(2000)
np.random.shuffle(index)
```

```python
train_X = train_X[index,:,:,:]
train_Y = train_Y[index]
```

```python
print(train_X.shape)
print(train_Y.shape)
```

```python
batch_size = 20
epochs = 20
```

```python
model.fit(train_X, train_Y, batch_size = batch_size, epochs=epochs)
```
